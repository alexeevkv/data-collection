# 1.1 Скачивание наборов данных Kaggle

```python
import kaggle
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', None)
sns.set_theme()
import warnings
warnings.filterwarnings("ignore")
```

Поиск наборов данных на Kaggle по ключевому слова большие данные - big data


```python
datasets = kaggle.api.datasets_list(search="big data")

print(datasets)
```

```python
data = pd.DataFrame(datasets)
data.head()
```

Для анализа будем использовать следующие колонки:
- title
- downloadCount
- viewCount
- topicCount
- kernelCount


```python
faetures = [
    'title',
    'downloadCount',
    'viewCount',
    'topicCount',
    'kernelCount',
]
```

```python
data = data[faetures]
```

```python
data
```

Посмотрим базовые статистики и самые популярные датасеты по скачиваниям и просмотрам

```python
print("Total Downloads:", data['downloadCount'].sum())
print("Total Views:", data['viewCount'].sum())
print("Average Downloads:", data['downloadCount'].mean())
print("Average Views:", data['viewCount'].mean())
```

```python
top_downloads = data.nlargest(5, 'downloadCount')
print("\nTop Datasets by Download Count:\n", top_downloads[['title', 'downloadCount']])
```

```python
top_views = data.nlargest(5, 'viewCount')
print("\nTop Datasets by View Count:\n", top_views[['title', 'viewCount']])
```

Посотрим на корреляции между загрузками, просмотрами и запусками

```python
correlation_downloads_views = data['downloadCount'].corr(data['viewCount'])
correlation_downloads_kernels = data['downloadCount'].corr(data['kernelCount'])
correlation_views_kernels = data['viewCount'].corr(data['kernelCount'])

print("\nCorrelation between Downloads and Views:", correlation_downloads_views)
print("Correlation between Downloads and Kernel Count:", correlation_downloads_kernels)
print("Correlation between Views and Kernel Count:", correlation_views_kernels)
```

Видим, что наибольшая корреляция есть между просмотрами и kernel count (это сколько раз был запущен ноутбук на kaggle). Это говорит о том, что очень часто люди после просмотра дополнительно запускают ноутбуки с целью лучше разобраться

Посмотрим теперь на распределения показателей (загрузки, просмотра и запуска), а также их совсем распределением

```python
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
sns.histplot(data['downloadCount'], bins=10, kde=True)
plt.title('Distribution of Download Counts')
plt.xlabel('Download Count')
plt.ylabel('Frequency')

plt.subplot(2, 2, 2)
sns.histplot(data['viewCount'], bins=10, kde=True)
plt.title('Distribution of View Counts')
plt.xlabel('View Count')
plt.ylabel('Frequency')

plt.subplot(2, 2, 3)
sns.histplot(data['kernelCount'], bins=10, kde=True)
plt.title('Distribution of Kernel Counts')
plt.xlabel('Kernel Count')
plt.ylabel('Frequency')

plt.subplot(2, 2, 4)
sns.scatterplot(data=data, x='downloadCount', y='viewCount', hue='kernelCount', size='kernelCount', sizes=(20, 200), alpha=0.7)
plt.title('Downloads vs Views (size by Kernel Count)')
plt.xlabel('Download Count')
plt.ylabel('View Count')

plt.tight_layout()
plt.show()
```

Можно увидеть, что большинство датасетов очень редко смотрят. Скорее всего это связано с positional bias а также с тем, что популярные датасета становятся более популярными, а не популярные так и остатся неизведанными. Видимо в выдачах на кагле никак не борются с expolarion vs exploitation  диллемой
