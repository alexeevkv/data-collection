# 1.1 Скачивание наборов данных Kaggle

```python
import kaggle
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', None)
sns.set_theme()
import warnings
warnings.filterwarnings("ignore")
```

```python
# поиск наборов данных на Kaggle о ""ключевое слово"""
datasets = kaggle.api.datasets_list(search="big data")

print(datasets)
```

```python
df = pd.DataFrame(datasets)
df.head()
```

Для анализа будем использовать следующие колонки:
- title
- downloadCount
- viewCount
- topicCount
- kernelCount


```python
faetures = [
    'title',
    'downloadCount',
    'viewCount',
    'topicCount',
    'kernelCount',
]
```

```python
data = df[faetures]
```

```python
data
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='title', y='downloadCount', palette='Blues')
plt.title('Download Counts by Dataset')
plt.xlabel('Dataset Title')
plt.ylabel('Download Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='title', y='viewCount', palette='Oranges')
plt.title('View Counts by Dataset')
plt.xlabel('Dataset Title')
plt.ylabel('View Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['title'], df['downloadCount'], marker='o', label='Downloads', color='blue')
plt.plot(df['title'], df['viewCount'], marker='o', label='Views', color='orange')
plt.title('Downloads and Views by Dataset')
plt.xlabel('Dataset Title')
plt.ylabel('Counts')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='downloadCount', y='viewCount', size='kernelCount', sizes=(50, 300), hue='topicCount', palette='viridis')
plt.title('Downloads vs. Views')
plt.xlabel('Download Count')
plt.ylabel('View Count')
plt.legend(title='Topic Count')
plt.show()
```

```python
sns.pairplot(df[['downloadCount', 'viewCount', 'topicCount', 'kernelCount']], diag_kind='kde', palette='husl')
plt.suptitle('Pair Plot of Dataset Usability Features', y=1.02)
plt.show()
```
