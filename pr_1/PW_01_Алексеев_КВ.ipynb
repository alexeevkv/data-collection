```python
import kaggle
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

ÐŸÐ¾Ð¸ÑÐº Ð½Ð°Ð±Ð¾Ñ€Ð¾Ð² Ð´Ð°Ð½Ð½Ñ‹Ñ… Ð½Ð° Kaggle Ð¿Ð¾ ÐºÐ»ÑŽÑ‡ÐµÐ²Ð¾Ð¼Ñƒ ÑÐ»Ð¾Ð²Ð° Ð±Ð¾Ð»ÑŒÑˆÐ¸Ðµ Ð´Ð°Ð½Ð½Ñ‹Ðµ - big data


```python
datasets = kaggle.api.datasets_list(search="big data")

print(datasets)
```

```python
data = pd.DataFrame(datasets)
data.head()
```

Ð”Ð»Ñ Ð°Ð½Ð°Ð»Ð¸Ð·Ð° Ð±ÑƒÐ´ÐµÐ¼ Ð¸ÑÐ¿Ð¾Ð»ÑŒÐ·Ð¾Ð²Ð°Ñ‚ÑŒ ÑÐ»ÐµÐ´ÑƒÑŽÑ‰Ð¸Ðµ ÐºÐ¾Ð»Ð¾Ð½ÐºÐ¸:
- title
- downloadCount
- viewCount

```python
faetures = [
    'title',
    'downloadCount',
    'viewCount',
    'topicCount',
    'kernelCount',
]
```

```python
data = data[faetures]
```

```python
data
```

ÐŸÐ¾ÑÐ¼Ð¾Ñ‚Ñ€Ð¸Ð¼ Ð±Ð°Ð·Ð¾Ð²Ñ‹Ðµ ÑÑ‚Ð°Ñ‚Ð¸ÑÑ‚Ð¸ÐºÐ¸ Ð¸ ÑÐ°Ð¼Ñ‹Ðµ Ð¿Ð¾Ð¿ÑƒÐ»ÑÑ€Ð½Ñ‹Ðµ Ð´Ð°Ñ‚Ð°ÑÐµÑ‚Ñ‹ Ð¿Ð¾ ÑÐºÐ°Ñ‡Ð¸Ð²Ð°Ð½Ð¸ÑÐ¼ Ð¸ Ð¿Ñ€Ð¾ÑÐ¼Ð¾Ñ‚Ñ€Ð°Ð¼

```python
print("Total Downloads:", data['downloadCount'].sum())
print("Total Views:", data['viewCount'].sum())
print("Average Downloads:", data['downloadCount'].mean())
print("Average Views:", data['viewCount'].mean())
```

```python
top_downloads = data.nlargest(5, 'downloadCount')
print("\nTop Datasets by Download Count:\n", top_downloads[['title', 'downloadCount']])
```

```python
top_views = data.nlargest(5, 'viewCount')
print("\nTop Datasets by View Count:\n", top_views[['title', 'viewCount']])
```

ÐŸÐ¾ÑÐ¾Ñ‚Ñ€Ð¸Ð¼ Ð½Ð° ÐºÐ¾Ñ€Ñ€ÐµÐ»ÑÑ†Ð¸Ð¸ Ð¼ÐµÐ¶Ð´Ñƒ Ð·Ð°Ð³Ñ€ÑƒÐ·ÐºÐ°Ð¼Ð¸, Ð¿Ñ€Ð¾ÑÐ¼Ð¾Ñ‚Ñ€Ð°Ð¼Ð¸ Ð¸ Ð·Ð°Ð¿ÑƒÑÐºÐ°Ð¼Ð¸

```python
correlation_downloads_views = data['downloadCount'].corr(data['viewCount'])
correlation_downloads_kernels = data['downloadCount'].corr(data['kernelCount'])
correlation_views_kernels = data['viewCount'].corr(data['kernelCount'])

print("\nCorrelation between Downloads and Views:", correlation_downloads_views)
print("Correlation between Downloads and Kernel Count:", correlation_downloads_kernels)
print("Correlation between Views and Kernel Count:", correlation_views_kernels)
```

```python
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
sns.histplot(data['downloadCount'], bins=10, kde=True)
plt.title('Distribution of Download Counts')
plt.xlabel('Download Count')
plt.ylabel('Frequency')

plt.subplot(2, 2, 2)
sns.histplot(data['viewCount'], bins=10, kde=True)
plt.title('Distribution of View Counts')
plt.xlabel('View Count')
plt.ylabel('Frequency')

plt.subplot(2, 2, 3)
sns.histplot(data['kernelCount'], bins=10, kde=True)
plt.title('Distribution of Kernel Counts')
plt.xlabel('Kernel Count')
plt.ylabel('Frequency')

plt.subplot(2, 2, 4)
sns.scatterplot(data=data, x='downloadCount', y='viewCount', hue='kernelCount', size='kernelCount', sizes=(20, 200), alpha=0.7)
plt.title('Downloads vs Views (size by Kernel Count)')
plt.xlabel('Download Count')
plt.ylabel('View Count')

plt.tight_layout()
plt.show()
```
