Данные сформируем самостоятельно

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
financial_data_2023 = {
  'Month': ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August',
 'September', 'October', 'November', 'December'],
  'Revenue': [100000, 120000, 115000, 130000, 125000, 135000, 140000, 150000,
 160000, 170000, 165000, 180000],
  'Expenses': [80000, 85000, 90000, 95000, 88000, 92000, 98000, 102000, 105000,
 110000, 108000, 115000],
  'Profit': [20000, 35000, 25000, 35000, 37000, 43000, 42000, 48000, 55000, 60000,
 57000, 65000]
 }
```

```python
data = pd.DataFrame.from_dict(financial_data_2023)
```

```python
total_revenue = data['Revenue'].sum()
total_expenses = data['Expenses'].sum()
total_profit = data['Profit'].sum()

print(f'Общий доход: {total_revenue:.2f}')
print(f'Общие расходы: {total_expenses:.2f}')
print(f'Общая прибыль: {total_profit:.2f}')
```

```python
# Построение графиков динамики финансовых потоков
plt.figure(figsize=(12, 6))

# График доходов
plt.plot(data['Month'], data['Revenue'], marker='o', label='Доходы', color='green')

# График расходов
plt.plot(data['Month'], data['Expenses'], marker='o', label='Расходы', color='red')

# График прибыли
plt.plot(data['Month'], data['Profit'], marker='o', label='Прибыль', color='blue')

plt.title('Динамика финансовых потоков по месяцам в 2023 году')
plt.xlabel('Месяц')
plt.ylabel('Сумма')
plt.xticks(rotation=45)
plt.legend()
plt.grid()

# Показать график
plt.tight_layout()
plt.show()
```

```python
data.to_csv('financial_data_2023.csv')
```
